### Barplot of studied hours and exam score
- It clearly shows that study hours is directly proportional to Exam Scores

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Bin Hours_Studied into intervals of 5
df['Study_Bin'] = pd.cut(df['Hours_Studied'], bins=range(0, df['Hours_Studied'].max()+5, 5))

# Plot average Exam_Score by Study_Bin
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Study_Bin', y='Exam_Score', estimator='mean')
plt.xticks(rotation=45)
plt.title('Average Exam Score by Hours Studied Interval')
plt.tight_layout()
plt.show()
```

### Imputing the missing values 

```python
df['Teacher_Quality'] = df['Teacher_Quality'].fillna(df['Teacher_Quality'].mode()[0])
```

```python
df['Parental_Education_Level'] = df['Parental_Education_Level'].fillna(df['Parental_Education_Level'].mode()[0])
```

```python

df['Distance_from_Home']= df['Distance_from_Home'].fillna(df['Distance_from_Home'].mode()[0])
```

## üîç Summary of Findings

After performing Exploratory Data Analysis (EDA) on the student performance dataset, several important patterns and relationships emerged:

### üìä Key Numeric Insights:
- **Attendance** shows the strongest positive correlation with exam performance (**r = 0.58**). Students with higher attendance tend to score significantly better, emphasizing the importance of classroom presence.
- **Hours Studied** also strongly impacts performance (**r = 0.45**), confirming that consistent study habits contribute meaningfully to academic success.
- **Previous Scores** and **Tutoring Sessions** have a moderate positive effect, suggesting that both prior knowledge and academic support play important roles.
- **Sleep Hours** and **Physical Activity**, while important for health, show very weak or no direct correlation with final exam scores in this dataset.

### üìä Key Categorical Insights:
- Students with **High Motivation Levels** consistently outperform their peers. Motivation appears to be a key psychological driver of success.
- Those with **Internet Access** tend to score higher, likely due to easier access to learning materials and online resources.
- A **Positive Peer Influence** is associated with better performance, suggesting the academic value of supportive social circles.
- **Parental Involvement** and **Teacher Quality** also show upward trends with Exam Scores, although with less pronounced effects.

### üìå Overall Conclusion:
Multiple interconnected factors contribute to student performance. Academic habits (like studying and attending class), motivational and environmental support (peer influence, internet, and parental involvement), and institutional support (tutoring, teaching quality) all play measurable roles.

Understanding these insights can help educators, parents, and students focus on the most impactful areas to drive improvement in academic outcomes.
